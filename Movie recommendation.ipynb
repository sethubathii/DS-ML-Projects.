in similarity_matrix 0 to 4806 are the movies and their similar movies.

It follows the below steps to make recommendations.

Get the index of the movie using the title. Get the list of similarity scores of the movies concerning all the movies. Enumerate them (create tuples) with the first element being the index and the second element is the cosine similarity score. Sort the list of tuples in descending order based on the similarity score. Get the list of the indices of the top 10 movies from the above sorted list. Exclude the first element because it is the title itself. Map those indices to their respective titles and return the movies list.

Now, let us create a series that maps the index of the matrix to movie names to make it easy for us to just feed in movie names and get the recommendation.

```python
#movie_index_mapping
mapping=pd.Series(movies_df.index,index=movies_df["title"])
mapping
```

```python
movie_index=mapping["John Carter"]
movie_index
```

```python
# create a function that will take the movie name and make recommendation

def recommend_movie(movie_input):
    movie_index = mapping[movie_input]
    # get similarity value with other movies
    similarity_score = list(enumerate(similarity[movie_index]))
    similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
    # get the list of top 10 recommended movies
    similarity_score = similarity_score[:10]

    movie_ind = [i[0] for i in similarity_score]
    recommended_movies = movies['title'].iloc[movie_ind]
    print('The recommended movies:', movie_input,"is:")
    return recommended_movies
```
